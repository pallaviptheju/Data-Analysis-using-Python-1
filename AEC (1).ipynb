6) Write a Python program to illustrate Stack and subplot usingMatplotlib.

```python
import matplotlib.pyplot as plt

# ---------- DATA ----------
days = [1, 2, 3, 4, 5]

sleeping = [7, 8, 6, 7, 8]
working = [8, 9, 7, 8, 9]
playing = [2, 3, 4, 3, 2]

# ---------- CREATING SUBPLOTS ----------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# ---------- STACK PLOT ----------
ax1.stackplot(days, sleeping, working, playing, labels=['Sleeping', 'Working', 'Playing'])
ax1.set_title("Daily Activity Stack Plot")
ax1.set_xlabel("Days")
ax1.set_ylabel("Hours")
ax1.legend(loc='upper left')

# ---------- SIMPLE LINE SUBPLOT ----------
ax2.plot(days, sleeping, label='Sleeping')
ax2.plot(days, working, label='Working')
ax2.plot(days, playing, label='Playing')
ax2.set_title("Subplot: Activity Lines")
ax2.set_xlabel("Days")
ax2.set_ylabel("Hours")
ax2.legend()

# Display both plots
plt.tight_layout()
plt.show()
```

7) Write a Python program which explains uses of customizing seaborn plots with Aesthetic functions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")

# -------------------------------
# 1. Set plot style
# -------------------------------
sns.set_style("whitegrid")      # other styles: darkgrid, white, dark, ticks

# -------------------------------
# 2. Set color palette
# -------------------------------
sns.set_palette("coolwarm")     # other palettes: deep, bright, muted, blues

# -------------------------------
# 3. Set plot context
# -------------------------------
sns.set_context("talk")         # other options: paper, notebook, poster

# -------------------------------
# Create a seaborn plot
# -------------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x="total_bill", y="tip", hue="time", style="sex", data=tips)

# Add title
plt.title("Customized Seaborn Scatter Plot with Aesthetic Functions")

# Show the plot
plt.show()
```

8)Write a Python program to explain working with bokeh line graph using Annotation sand Legends.

```python
from bokeh.plotting import figure, show
from bokeh.models import Span, Label
from bokeh.io import output_notebook

# Display in notebook (remove if running as script)
output_notebook()

# ----- Sample Data -----
x = [1, 2, 3, 4, 5]
y = [3, 7, 8, 5, 10]

# Create a figure
p = figure(title="Bokeh Line Graph with Annotations & Legend",
           x_axis_label="X Values",
           y_axis_label="Y Values",
           width=600,
           height=350)

# Draw line with legend
p.line(x, y, line_width=3, color="blue", legend_label="Trend Line")

# Add data points
p.circle(x, y, size=8, color="red", legend_label="Data Points")

# ----- Adding Annotations -----

# Add a horizontal line (annotation)
hline = Span(location=6, dimension='width', line_color='green', line_width=2, line_dash="dashed")
p.add_layout(hline)

# Add a text label annotation
label = Label(x=3, y=6.5, text="Average Level", text_color="green")
p.add_layout(label)

# Show legend on top-left
p.legend.location = "top_left"

# Display the plot
show(p)
```

9 Write a Python program for plotting different types of plots using Bokeh.

```python
from bokeh.plotting import figure, show
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource
from math import pi

# ---------------------------------
# 1. LINE PLOT
# ---------------------------------
line_fig = figure(title="Line Plot", width=350, height=300)
x = [1, 2, 3, 4, 5]
y = [3, 7, 5, 9, 6]
line_fig.line(x, y, line_width=2, color="blue")
line_fig.circle(x, y, size=8, color="red")


# ---------------------------------
# 2. SCATTER PLOT
# ---------------------------------
scatter_fig = figure(title="Scatter Plot", width=350, height=300)
scatter_fig.scatter([1,2,3,4,5], [5,4,6,7,3], size=10, color="green")


# ---------------------------------
# 3. BAR PLOT
# ---------------------------------
bar_fig = figure(title="Bar Plot", x_range=["Python", "Java", "C++", "AI"], width=350, height=300)
students = [40, 30, 20, 35]
bar_fig.vbar(x=["Python", "Java", "C++", "AI"], top=students, width=0.5, color="orange")


# ---------------------------------
# 4. PIE PLOT (Donut Style)
# ---------------------------------
from bokeh.transform import cumsum

data = {
    "courses": ["Python", "Java", "C++", "AI"],
    "count": [40, 30, 20, 10]
}
source = ColumnDataSource(data=data)

# Calculate angle
data['angle'] = [value / sum(data['count']) * 2 * pi for value in data['count']]
data['color'] = ["skyblue", "lightgreen", "pink", "gold"]

pie_fig = figure(title="Pie Chart", width=350, height=300, tools="hover", tooltips="@courses: @count")
pie_fig.wedge(x=0, y=1, radius=0.4,
              start_angle=cumsum("angle", include_zero=True),
              end_angle=cumsum("angle"),
              line_color="white",
              fill_color="color",
              legend_field="courses",
              source=data)


# ---------------------------------
# Arrange all plots in a grid
# ---------------------------------
grid = gridplot([[line_fig, scatter_fig],
                 [bar_fig, pie_fig]])

# Show the result
show(grid)
```

10 Write a Python program to draw 3D Plots using PlotlyLibraries.

```python
import plotly.graph_objects as go
import numpy as np

# -------------------------------
# 1. 3D LINE PLOT
# -------------------------------
x = np.linspace(0, 10, 50)
y = np.sin(x)
z = np.cos(x)

line_plot = go.Scatter3d(
    x=x, y=y, z=z,
    mode='lines',
    line=dict(width=5, color='blue'),
    name='3D Line'
)

# -------------------------------
# 2. 3D SCATTER PLOT
# -------------------------------
scatter_plot = go.Scatter3d(
    x=[1, 2, 3, 4],
    y=[2, 3, 1, 5],
    z=[5, 3, 6, 2],
    mode='markers',
    marker=dict(size=6, color='red'),
    name='3D Scatter'
)

# -------------------------------
# 3. 3D SURFACE PLOT
# -------------------------------
x_surface = np.linspace(-5, 5, 50)
y_surface = np.linspace(-5, 5, 50)
x_surface, y_surface = np.meshgrid(x_surface, y_surface)
z_surface = np.sin(np.sqrt(x_surface**2 + y_surface**2))

surface_plot = go.Surface(
    x=x_surface,
    y=y_surface,
    z=z_surface,
    colorscale='Viridis',
    name='3D Surface'
)

# -------------------------------
# Combine all plots
# -------------------------------
fig = go.Figure(data=[line_plot, scatter_plot, surface_plot])

fig.update_layout(
    title="3D Plot Examples Using Plotly",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    ),
    width=900,
    height=600
)

# Display all 3D plots
fig.show()
```

11)  Write a Python program to draw Time Series using Plotly Libraries.

```python
import plotly.graph_objects as go
import pandas as pd

# -------------------------------
# Sample time series data
# -------------------------------
dates = pd.date_range(start="2023-01-01", periods=10, freq='D')
values = [10, 12, 14, 11, 15, 18, 20, 17, 19, 22]

# -------------------------------
# Create time series figure
# -------------------------------
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=dates,
    y=values,
    mode='lines+markers',
    line=dict(width=2),
    marker=dict(size=6),
    name="Daily Value"
))

# -------------------------------
# Formatting layout
# -------------------------------
fig.update_layout(
    title="Time Series Plot Using Plotly",
    xaxis_title="Date",
    yaxis_title="Value",
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    width=800,
    height=450
)

# Show the plot
fig.show()
```

12) Write a Python program for creating Maps using Plotly Libraries. ( 7B)


```python
import plotly.express as px
import pandas as pd

# ------------------------------------
# Sample Data for World Choropleth Map
# ------------------------------------
data = {
    "country": ["India", "United States", "China", "Brazil", "Russia"],
    "value": [90, 75, 88, 65, 70]
}

df = pd.DataFrame(data)

# ------------------------------------
# 1. Choropleth World Map
# ------------------------------------
choropleth_map = px.choropleth(
    df,
    locations="country",
    locationmode="country names",
    color="value",
    color_continuous_scale="Viridis",
    title="Sample Choropleth Map (Country-wise Values)"
)

choropleth_map.show()


# ------------------------------------
# 2. Scatter Geo Map (Pinpoints on Map)
# ------------------------------------
scatter_data = {
    "city": ["Delhi", "New York", "Tokyo", "London"],
    "lat": [28.7041, 40.7128, 35.6895, 51.5074],
    "lon": [77.1025, -74.0060, 139.6917, -0.1278],
    "population": [19, 8, 37, 9]
}

df_scatter = pd.DataFrame(scatter_data)

scatter_map = px.scatter_geo(
    df_scatter,
    lat="lat",
    lon="lon",
    hover_name="city",
    size="population",
    projection="natural earth",
    title="Scatter Geo Map (Cities around the World)"
)

scatter_map.show()
```
